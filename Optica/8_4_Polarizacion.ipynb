# Polarización


1) LINEAL:
   El campo eléctrico oscila en una única dirección.
   Ej: E = E0 cos(ωt) î


```python
# ============================================
# POLARIZACIÓN DE ONDAS ELECTROMAGNÉTICAS
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------
# Parámetros de la onda
# --------------------------------------------
E0 = 1          # amplitud
k = 2*np.pi     # número de onda (normalizado)
omega = 2*np.pi # frecuencia angular (normalizada)

# eje de propagación (z) o tiempo equivalente
t = np.linspace(0, 4*np.pi, 400)

# ============================================
# 1. POLARIZACIÓN LINEAL
# E = E0 (cos(ωt), 0)
# ============================================

Ex_lin = E0 * np.cos(t)
Ey_lin = np.zeros_like(t)

# ============================================
# 2. POLARIZACIÓN CIRCULAR (antihoraria)
# E = E0 (cos(ωt), sin(ωt))
# ============================================

Ex_circ_ccw = E0 * np.cos(t)
Ey_circ_ccw = E0 * np.sin(t)

# ============================================
# 3. POLARIZACIÓN CIRCULAR (horaria)
# E = E0 (cos(ωt), -sin(ωt))
# ============================================

Ex_circ_cw = E0 * np.cos(t)
Ey_circ_cw = -E0 * np.sin(t)

# ============================================
# GRÁFICAS
# ============================================

fig = plt.figure(figsize=(15, 4))

# ---------------- LINEAL ----------------
ax1 = fig.add_subplot(1, 3, 1)
ax1.plot(Ex_lin, Ey_lin)
ax1.set_title("Polarización lineal")
ax1.set_xlabel("Ex")
ax1.set_ylabel("Ey")
ax1.axhline(0, color='gray', linewidth=0.5)
ax1.axvline(0, color='gray', linewidth=0.5)
ax1.set_aspect('equal')

# ---------------- CIRCULAR CCW ----------------
ax2 = fig.add_subplot(1, 3, 2)
ax2.plot(Ex_circ_ccw, Ey_circ_ccw)
ax2.set_title("Polarización circular (antihoraria)")
ax2.set_xlabel("Ex")
ax2.set_ylabel("Ey")
ax2.axhline(0, color='gray', linewidth=0.5)
ax2.axvline(0, color='gray', linewidth=0.5)
ax2.set_aspect('equal')

# ---------------- CIRCULAR CW ----------------
ax3 = fig.add_subplot(1, 3, 3)
ax3.plot(Ex_circ_cw, Ey_circ_cw)
ax3.set_title("Polarización circular (horaria)")
ax3.set_xlabel("Ex")
ax3.set_ylabel("Ey")
ax3.axhline(0, color='gray', linewidth=0.5)
ax3.axvline(0, color='gray', linewidth=0.5)
ax3.set_aspect('equal')

plt.tight_layout()
plt.show()

# ============================================
# REPRESENTACIÓN 3D: evolución en propagación
# (trayectoria del extremo del campo eléctrico)
# ============================================

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 4))

# Lineal en 3D
ax = fig.add_subplot(131, projection='3d')
ax.plot(t, Ex_lin, Ey_lin)
ax.set_title("Lineal (E vs propagación)")
ax.set_xlabel("z o t")
ax.set_ylabel("Ex")
ax.set_zlabel("Ey")

# Circular CCW
ax = fig.add_subplot(132, projection='3d')
ax.plot(t, Ex_circ_ccw, Ey_circ_ccw)
ax.set_title("Circular antihoraria")
ax.set_xlabel("z o t")
ax.set_ylabel("Ex")
ax.set_zlabel("Ey")

# Circular CW
ax = fig.add_subplot(133, projection='3d')
ax.plot(t, Ex_circ_cw, Ey_circ_cw)
ax.set_title("Circular horaria")
ax.set_xlabel("z o t")
ax.set_ylabel("Ex")
ax.set_zlabel("Ey")

plt.tight_layout()
plt.show()

# ============================================
# RESUMEN TEÓRICO (Markdown en Colab)
# ============================================

print("""
TIPOS DE POLARIZACIÓN:

1) LINEAL:
   El campo eléctrico oscila en una única dirección.
   Ej: E = E0 cos(ωt) î

2) CIRCULAR:
   Dos componentes ortogonales con desfase π/2:
   E = E0 [cos(ωt) î ± sin(ωt) ĵ]

   - signo + : antihoraria
   - signo - : horaria

Interpretación geométrica:
El extremo del vector E describe:
- Línea → polarización lineal
- Circunferencia → polarización circular
""")
```
