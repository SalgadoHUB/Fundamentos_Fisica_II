# Interferencias de ondas

La interferencia es el fenómeno que ocurre cuando dos o más ondas se superponen en el espacio. El resultado es una nueva onda cuya amplitud depende de la suma algebraica de las ondas individuales.

---

## Ondas sinusoidales

Consideremos dos ondas de igual amplitud $A$, igual frecuencia angular $\omega$, y mismo número de onda $k$, pero con distinta fase:

$
y_1(x,t) = A \sin(kx - \omega t)
$

$
y_2(x,t) = A \sin(kx - \omega t + \phi)
$

donde $\phi$ es la **diferencia de fase**. La diferencia de fase $\phi\$ mide el desplazamiento relativo entre dos ondas.

- $\phi = 0$: interferencia constructiva perfecta  
- $\phi = \pi: $interferencia destructiva total  
- $0 < \phi < \pi$: interferencia parcial  

---

## Superposición de ondas

La onda resultante es:

$
y = y_1 + y_2
$

Sustituyendo:

$
y = A \sin(kx - \omega t) + A \sin(kx - \omega t + \phi)
$

Usando la identidad trigonométrica:

$
\sin a + \sin b = 2 \sin\left(\frac{a+b}{2}\right)\cos\left(\frac{a-b}{2}\right)
$

obtenemos:

$
y = 2A \cos\left(\frac{\phi}{2}\right)\sin\left(kx - \omega t + \frac{\phi}{2}\right)
$

La amplitud de la onda resultante es:

$
A_{\text{res}} = 2A \cos\left(\frac{\phi}{2}\right)
$

Esto muestra que la interferencia depende únicamente de la diferencia de fase. Así, por ejemplo, tenemos:

- Si $\phi = 0$, entonces $A_{\text{res}} = 2A$ → máxima interferencia constructiva  
- Si $\phi = \pi$, entonces $A_{\text{res}} = 0$ → cancelación total  
- Si $\phi = \frac{\pi}{2}$, interferencia parcial  



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
A = 1
k = 2 * np.pi
omega = 2 * np.pi
t = 0  # instante fijo

x = np.linspace(0, 4, 1000)

# Diferentes fases
phi_values = [0, np.pi/2, np.pi]

plt.figure(figsize=(12, 6))

for i, phi in enumerate(phi_values):
    y1 = A * np.sin(k*x - omega*t)
    y2 = A * np.sin(k*x - omega*t + phi)
    y = y1 + y2

    plt.subplot(3, 1, i+1)
    plt.plot(x, y1, '--', label='Onda 1')
    plt.plot(x, y2, '--', label='Onda 2')
    plt.plot(x, y, label='Resultado', linewidth=2)

    plt.title(f'Interferencia con φ = {phi:.2f} rad')
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()
```

# Gráfica con control del desfase

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Parámetros de las ondas
A = 1
k = 2 * np.pi
omega = 2 * np.pi
t = 0

x = np.linspace(0, 4, 1000)

# Función de actualización
def plot_interferencia(phi):
    y1 = A * np.sin(k*x - omega*t)
    y2 = A * np.sin(k*x - omega*t + phi)
    y = y1 + y2

    plt.figure(figsize=(10, 5))
    plt.plot(x, y1, '--', label='Onda 1')
    plt.plot(x, y2, '--', label='Onda 2')
    plt.plot(x, y, linewidth=2, label='Resultado')

    plt.title(f'Interferencia de ondas (φ = {phi:.2f} rad)')
    plt.xlabel('x')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()
    plt.ylim(-2.2, 2.2)
    plt.show()

# Slider interactivo
interact(
    plot_interferencia,
    phi=FloatSlider(
        value=0,
        min=0,
        max=2*np.pi,
        step=0.05,
        description='Fase φ'
    )
)
```

# Interferencias de dos ondas armónicas en 2D

Consideremos dos fuentes puntuales coherentes (dos focos) en un medio bidimensional. Ambas emiten ondas armónicas con:

- Igual amplitud $A$
- Igual frecuencia angular $\omega$
- Misma longitud de onda $\lambda$
- Diferencia de fase inicial nula (fuentes coherentes)

Estas ondas se propagan en todas direcciones en el plano.

---

## Ecuación de las ondas en 2D

Si los focos están en posiciones $\mathbf{r}_1$ y $\mathbf{r}_2$, la perturbación en un punto $\mathbf{r}$ es:

$
y_1(\mathbf{r}, t) = \frac{A}{r_1}\sin(k r_1 - \omega t)
$

$
y_2(\mathbf{r}, t) = \frac{A}{r_2}\sin(k r_2 - \omega t)
$

donde:

- $r_1 = |\mathbf{r} - \mathbf{r}_1|$
- $r_2 = |\mathbf{r} - \mathbf{r}_2|$
- $k = \frac{2\pi}{\lambda}$

---

## Superposición

Por el principio de superposición:

$
y(\mathbf{r}, t) = y_1 + y_2
$

---

## Diferencia de fase espacial

La clave del fenómeno es la **diferencia de camino óptico**:

$
\Delta r = r_2 - r_1
$

Esto genera una diferencia de fase:

$
\Delta \phi = k (r_2 - r_1)
$

---

## Condiciones de interferencia

### Interferencia constructiva

Ocurre cuando las ondas llegan en fase:

$
\Delta r = m\lambda \quad (m \in \mathbb{Z})
$

$
\Delta \phi = 2\pi m
$

→ Se observan **máximos de intensidad** (franjas brillantes)

---

### Interferencia destructiva

Ocurre cuando llegan en oposición de fase:

$
\Delta r = \left(m + \frac{1}{2}\right)\lambda
$

$
\Delta \phi = (2m+1)\pi
$

→ Se observan **mínimos de intensidad** (zonas oscuras)

---

## Patrón de interferencia

La superposición de dos ondas en 2D genera un patrón característico de:

- Franjas curvas
- Hipérbolas de máximos y mínimos
- Simetría respecto al eje entre los dos focos

Esto ocurre porque el conjunto de puntos con diferencia de distancia constante define una **hipérbola**.

---

## Intensidad resultante

La intensidad observada es proporcional al cuadrado de la amplitud:

$
I \propto |y|^2
$

Para dos fuentes idénticas:

$
I(\mathbf{r}) \propto 2A^2 \left[1 + \cos(k(r_2 - r_1))\right]
$

---

Así que:

- Donde las ondas llegan en fase → máxima energía
- Donde llegan en oposición → cancelación
- El patrón es estacionario si las fuentes son coherentes

Este fenómeno aparece en:

- Ondas en agua (dos vibradores)
- Interferencia de luz (experimento de Young)
- Antenas emisoras
- Acústica en salas



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# -----------------------------
# Parámetros físicos
# -----------------------------
A = 1
k = 2 * np.pi     # lambda = 1
omega = 2 * np.pi
t_step = 0.05

# -----------------------------
# Malla 2D
# -----------------------------
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)

# Posición de los dos focos
x1, y1 = -1.5, 0
x2, y2 =  1.5, 0

# Distancias a cada fuente
r1 = np.sqrt((X - x1)**2 + (Y - y1)**2)
r2 = np.sqrt((X - x2)**2 + (Y - y2)**2)

# -----------------------------
# Figura
# -----------------------------
fig, ax = plt.subplots(figsize=(6, 6))
img = ax.imshow(np.zeros_like(X), extent=[-5,5,-5,5], origin='lower', cmap='RdBu')

ax.set_title("Interferencia de dos fuentes coherentes")
ax.set_xlabel("x")
ax.set_ylabel("y")

# -----------------------------
# Función de actualización
# -----------------------------
def update(frame):
    t = frame * t_step

    # Ondas individuales
    y1 = (A / np.sqrt(r1 + 1e-6)) * np.sin(k * r1 - omega * t)
    y2 = (A / np.sqrt(r2 + 1e-6)) * np.sin(k * r2 - omega * t)

    # Superposición
    y = y1 + y2

    img.set_data(y)
    img.set_clim(-2, 2)

    return [img]

# -----------------------------
# Animación
# -----------------------------
ani = FuncAnimation(fig, update, frames=200, interval=30, blit=True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Parámetros físicos
lambda_ = 500e-9   # longitud de onda (500 nm)
L = 1.0            # distancia a la pantalla (m)

# Eje en la pantalla
x = np.linspace(-0.01, 0.01, 1000)

def intensidad(x, d):
    beta = (np.pi * d * x) / (lambda_ * L)
    return np.cos(beta)**2

# Función para graficar
def plot_interferencia(d):
    plt.figure(figsize=(8,4))
    plt.plot(x, intensidad(x, d))
    plt.title(f"Interferencia de doble rendija (d = {d:.2e} m)")
    plt.xlabel("Posición en la pantalla (m)")
    plt.ylabel("Intensidad")
    plt.ylim(0,1.1)
    plt.grid()
    plt.show()

# Slider interactivo
interact(
    plot_interferencia,
    d=FloatSlider(value=1e-4, min=1e-5, max=5e-4, step=1e-5, description='d (m)')
);
```
