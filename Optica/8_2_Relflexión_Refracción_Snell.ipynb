# Reflexión y refracción de la luz

## Ley de la reflexión

Cuando un rayo de luz incide sobre una superficie, se cumple:

$$
\theta_i = \theta_r
$$

donde:
- $\theta_i$ es el ángulo de incidencia
- $\theta_r$ es el ángulo de reflexión

Ambos ángulos se miden respecto a la normal a la superficie.

---

## Ley de la refracción (Ley de Snell)

Cuando la luz pasa de un medio a otro con distinta velocidad, cambia de dirección:

$$
n_1 \sin \theta_1 = n_2 \sin \theta_2
$$

donde:
- $n_1, n_2$ son los índices de refracción
- $\theta_1$ es el ángulo de incidencia
- $\theta_2$ es el ángulo de refracción

---

## Interpretación física

El índice de refracción está relacionado con la velocidad de la luz:

$$
n = \frac{c}{v}
$$

Por tanto:

- Si $v_2 < v_1$ → la luz se acerca a la normal
- Si $v_2 > v_1$ → la luz se aleja de la normal

---

## Condición de reflexión total interna

Si $n_1 > n_2$, existe un ángulo crítico:

$$
\theta_c = \arcsin\left(\frac{n_2}{n_1}\right)
$$

Para $\theta_1 > \theta_c$, no hay refracción, solo reflexión.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def snell(theta1_deg=30, n1=1.0, n2=1.5):
    theta1 = np.radians(theta1_deg)

    # Ley de Snell
    sin_theta2 = (n1/n2) * np.sin(theta1)

    fig, ax = plt.subplots(figsize=(6,6))

    # Normal
    ax.plot([0,0], [-1,1], linestyle='--')

    # Interfaz
    ax.plot([-1,1], [0,0])

    # --------- RAYO INCIDENTE ----------
    x1 = -np.sin(theta1)
    y1 = np.cos(theta1)
    ax.plot([0, x1], [0, y1])

    # --------- RAYO REFLEJADO ----------
    ax.plot([0, -x1], [0, y1])

    # --------- ARCO θ1 ----------
    arc_theta = np.linspace(0, theta1, 50)
    ax.plot(np.sin(arc_theta)*0.3, np.cos(arc_theta)*0.3)

    # --------- REFRACCIÓN ----------
    if abs(sin_theta2) <= 1:
        theta2 = np.arcsin(sin_theta2)

        x2 = np.sin(theta2)
        y2 = -np.cos(theta2)
        ax.plot([0, x2], [0, y2])

        # arco θ2
        arc_theta2 = np.linspace(0, theta2, 50)
        ax.plot(np.sin(arc_theta2)*0.3, -np.cos(arc_theta2)*0.3)

        title = f"θ1 = {theta1_deg:.1f}°, θ2 = {np.degrees(theta2):.1f}°"
    else:
        title = "Reflexión total interna"

    # --------- TEXTO MEDIOS ----------
    ax.text(-0.9, 0.7, f"n1 = {n1}")
    ax.text(-0.9, -0.8, f"n2 = {n2}")

    ax.set_title(title)
    ax.set_xlim(-1,1)
    ax.set_ylim(-1,1)
    ax.set_aspect('equal')
    ax.grid()

    plt.show()

interact(
    snell,
    theta1_deg=(0, 89, 1),
    n1=(1.0, 2.5, 0.1),
    n2=(1.0, 2.5, 0.1)
)
```
